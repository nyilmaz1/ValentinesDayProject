```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Messages.csv', encoding = 'utf-8')
data.head()
```

```python
df = pd.DataFrame(data)
```

```python
#how many texts we sent:
received_count = df.Received.count()
sent_count = df.Sent.count()
print(received_count)
print(sent_count)
```

```python
#how many times he said love:
a = df.Received.str.count('love').sum()
b = df.Received.str.count('Love').sum()
c = df.Received.str.count('LOVE').sum()
received_count_love = a+b+c
print(received_count_love)

#how many times I said love:
d = df.Sent.str.count('love').sum()
e = df.Sent.str.count('Love').sum()
f = df.Sent.str.count('LOVE').sum()
sent_count_love = d+e+f
print(sent_count_love)


#how many times he said baby:
g = df.Received.str.count('baby').sum()
h = df.Received.str.count('Baby').sum()
i = df.Received.str.count('BABy').sum()
received_count_baby = int(g+h+i)
print(received_count_baby)


#how many times I said baby:
j = df.Sent.str.count('baby').sum()
k = df.Sent.str.count('Baby').sum()
l = df.Sent.str.count('BABY').sum()
sent_count_baby =int(j+k+l)
print(sent_count_baby)
```

```python
#add weekdays to the dataframe
weekdays = df.Date.dt.weekday
weekdays
df = df.assign(weekdays=weekdays)
df.head()
```

```python
#replace integers with actual values:
weekday_names = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
weekday_dict = {key: weekday_names[key] for key in range(7)}

df.weekdays = df.weekdays.apply(lambda x: weekday_dict[x])
```

```python
df.head(n=15)
```

```python
#let's group by these weekdays
weekday_counts = df.groupby('weekdays').count()
weekday_counts = weekday_counts.loc[weekday_names]
weekday_counts
```

```python
#let's plot this


weekday_counts.Date.plot(kind = "bar", title="Texts sent on each weekday", figsize = (7,3))
```
